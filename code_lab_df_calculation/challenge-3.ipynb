```python
grupo = orders.groupby('CustomerID').sum()

grupo.head()
```

```python
per_95 = grupo.amount_spent.quantile(0.95)  #numero

per_75 = grupo.amount_spent.quantile(0.75)
```

## Q2: How to identify which country has the most VIP Customers?

```python
# identificamos los vip

grupo[grupo.amount_spent >= per_95]  # VIP

vip = grupo[grupo.amount_spent >= per_95].index

vip
```

```python
orders[orders.CustomerID.isin(vip)].Country.value_counts()
```

```python
orders[orders.CustomerID.isin(vip)].drop_duplicates(subset=['CustomerID', 'Country'])
```

```python
orders[orders.CustomerID.isin(vip)].drop_duplicates(subset=['CustomerID']).Country.value_counts()
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# Identificamos a los clientes preferentes

grupo[(grupo.amount_spent <= per_95) & (grupo.amount_spent >= per_75)]

otros = grupo[(grupo.amount_spent <= per_95) & (grupo.amount_spent >= per_75)].index

otros  # CustomerID
```

```python
# queremos preferentes y vip 

todos = grupo[grupo.amount_spent >= per_75].index

todos  # CustomerID
```

```python
orders[orders.CustomerID.isin(todos)].drop_duplicates(subset=['CustomerID']).Country.value_counts()
```
