```python
# your code here
import pandas as pd
import numpy as np
data = pd.read_csv('../your-code/Warehouse_and_Retail_Sales_20240205.csv')
data.head(20)
```

```python
data.columns = [e.replace(' ', '_').lower() for e in data]
data.head(10)
```

```python
data.supplier = data.supplier.fillna('otro')
```

```python
data = data.dropna()
```

```python
data.head(10)
```

```python
data.isna().sum()
```

```python
data.to_csv('Clean data.csv', index = False)
```

```python
data_by_supplier = data.groupby('supplier', as_index = False).aggregate({'year': 'first', 'month' : 'first', 'item_type': 'first','retail_sales':'sum', 'retail_transfers': 'sum','warehouse_sales': 'sum'})
```

```python
data_by_supplier
```

```python
data_by_supplier.to_csv('Data by supplier.csv', index = False)
```

```python
data_by_items = data.groupby('item_type', as_index = False).aggregate({'year': 'first', 'month' : 'first', 'item_type': 'first', 'retail_sales':'sum', 'retail_transfers': 'sum','warehouse_sales': 'sum'})
```

```python
data_by_items
```

```python
data.to_csv('Data by items.csv', index = False)
```
