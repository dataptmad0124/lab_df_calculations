```python
def moda(lst):
    
    lst = lst.to_list()
    
    letra = ''
    
    num = 0
    
    for e in lst:
        
        if lst.count(e) > num:
            
            letra = e
            num = lst.count(e)
        
    
    return letra
```

```python
df.groupby('SUPPLIER').agg({'ITEM TYPE': moda}).head()
```
